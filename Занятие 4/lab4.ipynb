# Замена пропущенных данных
Возраст - важная характеристика для выживания.
Нужны способы замены пропущенных значений (метод.replace)

1 способ (грубый): замена на среднее или медиану (см. метод .describe())

```python
df.Age.describe()
```

```python
old_age=df.Age.copy()
df.Age=df.Age.replace(np.nan,28)
print(df.isna().sum())
df
```

Восстановите возрост

Проанализируйте имена пассажиров и выделите несколько групп ("Mr", "Mrs", "Miss", и т.п.) Известно,что статус человека, отражённый в имени, тесно коррелирует с возрастом, и эту взаимосвязь можно использовать для определения возраста пассажиров, возраст которых в исходных данных пропущен

Проведите анализ выживаемости в каждой группе. Вычислите медианы и выполните замены пропусков по возросту в каждой группе

```python
df.Age=old_age
print(df.isna().sum(), end = "\n"*2)

mean_age_by_category = df.groupby(df['Name'].apply(lambda x: x.split(',')[1].split()[0]))['Age'].mean()
part_names = mean_age_by_category.index
mean_ages = mean_age_by_category.values

plt.figure(figsize=(10, 7))
plt.barh(part_names, mean_ages)
plt.ylabel('Части имен')
plt.xlabel('Средний возраст')
plt.title('Средний возраст по частям имен')
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()

for part_name, mean_age in mean_age_by_category.items():
    df.loc[df['Name'].str.contains(part_name) & df['Age'].isna(), 'Age'] = mean_age
print(df.isna().sum())
```

Номер каюты

Среди номеров каюты пропущены данные для 80% пассажиров. Восстановить эти данные не представляется возможным. Для тех же пассажиров, для которых номер каюты известен, можно было бы извлечь номер палубы и на каком борту была каюта (буква в номере соответствует палубе, нечетные номера соответствуют левому борту).
Сведения о каютах пассажиров стали известны благодаря списку, найденному на теле погибшего стюарта Герберта Кейва, причем в список были включены только пассажиры первого класса (больше лифтов, стюартов, ближе к шлюпкам). Это означает, что значение может иметь само наличие или отсутствие данных о каюте пассажира.

Выполните анализ выживаемости в зависимости от наличия номера каюты.
Выполните замену категориального признака на бинарный (есть или нет номера)

```python
df['HasCabin'] = np.where(df['Cabin'].isna(), 0, 1)
print(df['HasCabin'])
survival_by_cabin = df.groupby('HasCabin')['Survived'].mean()
print(survival_by_cabin)

plt.figure(figsize=(10, 5))
survival_by_cabin.plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Наличие каюты')
plt.ylabel('Выживаемость')
plt.title('Ввыживаемость в зависимости от наличия номера каюты')
plt.xticks([0, 1], ['Нет', 'Есть'], rotation=0)
plt.grid(True)
plt.show()
```

Может заменим билеты с нулевой стоимостью на медианную стоимость для соответствующего класса?

```python
old_fare=df.Fare.copy()
```

```python
df.Fare = old_fare
print((df['Fare'] == 0).sum())

median_fares_by_class = df.groupby('Pclass')['Fare'].median()
classes = median_fares_by_class.index
prices = median_fares_by_class.values
for pclass, median_fare in median_fares_by_class.items():
    df.loc[(df['Fare'] == 0) & (df['Pclass'] == pclass), 'Fare'] = median_fare
print((df['Fare'] == 0).sum())

plt.figure(figsize=(10, 7))
plt.barh(classes, prices)
plt.ylabel('Классы кают')
plt.xlabel('Стоимость билета')
plt.title('Средняя стоимость билета по классе')
plt.grid(True)
plt.show()
```
